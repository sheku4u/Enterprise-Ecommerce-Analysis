```python
# removing outlierfrom sales 
q1 = orders['sales'].quantile(0.25)
q3 = orders['sales'].quantile(0.75)

iqr = q3 - q1

lower_sales = q1 - 1.5  * iqr
high_sales = q3 + 1.5  * iqr

sales_outlier = orders[
    (orders['sales'] < lower_sales) |  
    (orders['sales'] > high_sales)
]


# remove 
orders = orders[
    (orders['sales'] > lower_sales) & 
    (orders['sales'] < high_sales)
] 
```

```python
orders = orders[orders['sales'] > 0]
```
